```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("../Data/Cleaned/hr_cleaned.csv")
df.head()
```

```python
#what is attrition rate
df['Attrition'].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Attrition',
    data=df
)
plt.title("Attrition Distribution of Employees")
plt.show()
```

```python
plt.savefig("../Screenshots/attrition_distribution_of_employees.png",bbox_inches='tight')
```

```python
#which department has highest attrition
plt.figure(figsize=(10,5))
sns.countplot(x='Department',hue='Attrition',data=df)
plt.xticks(rotation=0)
plt.title("department Wise Attrition")
plt.show()
plt.savefig("../Screenshots/department wise attrition.png",bbox_inches='tight')
```

```python
#salary vs attrition
avg_salary = df.groupby('Attrition')['MonthlyIncome'].mean().reset_index()
plt.figure(figsize=(4,4))
sns.barplot(x='Attrition', y='MonthlyIncome', data=avg_salary)
plt.title("Average Salary vs Attrition")
plt.ylabel("Average Monthly Income")
plt.show()
plt.savefig("../Screenshots/Average salary vs Attrition.png",bbox_inches='tight')
```

```python
#overtime vs attrition
attrition_rate = pd.crosstab(df['OverTime'],df['Attrition'],normalize='index') * 100
attrition_rate.plot(kind='bar',figsize=(8,5))
plt.title("Attrition Rate by Overtime")
plt.ylabel("Percentage")
plt.show()
plt.savefig("../Screenshots/attrition vs Overtime.png",bbox_inches='tight')
```

```python
#job satisfaction vs attrition
plt.figure(figsize=(8,5))
sns.countplot(x='JobSatisfaction',hue='Attrition',data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
plt.savefig("../Screenshots/job satisfaction vs Attrition.png",bbox_inches='tight')
```

```python
#age group analysis
plt.figure(figsize=(8,5))
sns.countplot(x='age_group',hue='Attrition',data=df)
plt.title("Age Group vs Attrition")
plt.show()
plt.savefig("../Screenshots/age group vs attrition.png",bbox_inches='tight')
```
