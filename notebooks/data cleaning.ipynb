```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../Data/Raw/New HR_Attrition_dataset.csv",low_memory=False)
df.head()
```

Basic inspection or reading of data

```python
df.shape
df.info()
```

checking missing values


```python
df.isnull().sum()
```

```python
print(df['HourlyRate'].isnull().sum())
df['HourlyRate'].describe()
```

```python
df['HourlyRate']=pd.to_numeric(df['HourlyRate'],errors='coerce')
```

```python
df['HourlyRate'].isnull().sum()
df['HourlyRate'].describe()
```

```python
#filling the missing value
df['HourlyRate'].fillna(df['HourlyRate'].median(),inplace=True)
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
#feature engineering
df['AnnualIncome']=df['MonthlyIncome']*12
# print(df['AnnualIncome'])
```

```python
df['Attritionsign']=df['Attrition'].map({'Yes':1,'No':0})
```

```python
df['overtimedone']=df['OverTime'].map({'Yes':1,'No':0})
```

```python
bins=[18,25,35,45,55,65]
labels=["18-25","26-35","36-45","46-55","56-65"]
df["age_group"]=pd.cut(df["Age"],
                       bins=bins,
                       labels=labels)
```

Save the cleaned Dataset

```python
df.to_csv("../Data/Cleaned/hr_cleaned.csv",index=False)
```
