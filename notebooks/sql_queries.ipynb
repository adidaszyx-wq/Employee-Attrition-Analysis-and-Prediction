```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv("../Data/Cleaned/hr_cleaned.csv")
conn = sqlite3.connect("../mySql/hr_analytics.db")
df.to_sql("employee_data",conn,if_exists="replace",index=False
)
```

```python
query=""" SELECT COUNT(*) AS total_employees
FROM employee_data
GROUP BY Attrition;
"""
pd.read_sql_query(query,conn)#test sql table
```

```python
query = """
SELECT ROUND(SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*),2) AS attrition_rate
FROM employee_data;
"""
pd.read_sql_query(query, conn)#attrition rate
```

```python
query = """
SELECT Department, COUNT(*) AS attrition_count
FROM employee_data
WHERE Attrition = 'Yes'
GROUP BY Department
ORDER BY attrition_count DESC;
"""

pd.read_sql_query(query, conn)#Department-wise attrition
```

```python
query = """
SELECT OverTime, Attrition, COUNT(*) AS total
FROM employee_data
GROUP BY OverTime, Attrition
ORDER BY OverTime, Attrition;
"""

pd.read_sql_query(query, conn)#Overtime impact
```

```python
query = """
SELECT JobRole, ROUND(AVG(MonthlyIncome), 2) AS avg_salary
FROM employee_data
GROUP BY JobRole
ORDER BY avg_salary DESC;
"""

pd.read_sql_query(query, conn)#Average salary by job role
```
